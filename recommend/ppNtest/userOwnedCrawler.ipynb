```
#유저 정보 크롤링

gamedatas = pd.DataFrame(columns=['appid',	'playtime_forever',	'playtime_2weeks',	'steamid'])
# firstUser = 76561197960265744 + 65000
firstUser = 76561197961550000
availableUserCount = 0

for userid in range(firstUser, firstUser + 6100):

  url = 'https://api.steampowered.com/IPlayerService/GetOwnedGames/v1/?key=F1DAAB7BF3894FBDC83881C81C20803A&steamid=' + str(userid)
  response = requests.get(url)


  if response.status_code == 200:
      html = response.text
      soup = BeautifulSoup(html, 'html.parser') 
      rawdata = pd.read_json(soup.text)
      if rawdata.empty:
        continue
      elif rawdata['response']['game_count'] == 0:
        continue
  else :
    print(response.status_code)

  availableUserCount = availableUserCount + 1
  print(userid, availableUserCount)
  gamedata = pd.DataFrame(rawdata['response']['games'])
  gamedata = gamedata.drop(['playtime_windows_forever', 'playtime_mac_forever', 'playtime_linux_forever'], axis='columns')
  gamedata['steamid'] = userid
  # print(gamedata)
  gamedatas = pd.concat([gamedatas, gamedata] , ignore_index=True)

print(availableUserCount)
gamedatas
```
