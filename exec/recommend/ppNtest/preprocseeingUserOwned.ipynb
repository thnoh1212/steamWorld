```
#weight 할당 def
def nonePlaying(playtime_forever, playtime_2weeks, weight):
  #플레이시간 0
  if playtime_forever == float('-inf'):
    weight = 1
  #플레이시간 1시간 이내
  elif playtime_forever < 0:
    weight = -1
  #플레이시간 1시간 이상
  else :
    weight = 2

  if playtime_2weeks >= 60:
    weight = weight + 1
  return weight
```

```
#weight 추가
rawData['weight'] = 0
rawData.head(20)
```

```
#전체 데이터에 weight 부여

users = rawData
users['playtime_forever'] = users['playtime_forever']/60
users['playtime_forever'] = np.log10(users['playtime_forever'])
users['weight'] = users.apply(lambda x : nonePlaying(x['playtime_forever'], x['playtime_2weeks'], x['weight']), axis = 1)
users
```

```
#저장된 데이터 확인

dataConfrim = pd.read_csv("google_drive/steamData/userOwnedResult.csv")
dataConfrim = dataConfrim.drop(columns=["Unnamed: 0"])
dataConfrim
```

```
#user index 설정
noneDuplicate = noneDuplicate.sort_values(by = ['steamid'], ascending=True)
noneDuplicate['newsteamid'] = 0
```

```
#index 재설정
noneDuplicate.index = range(0, noneDuplicate.shape[0])
noneDuplicate
```

```
#id 변환

for i in noneDuplicate.index:
  if i == 0:
    counter = 0
  elif noneDuplicate['steamid'][i-1] != noneDuplicate['steamid'][i]:
    counter = counter + 1   
  noneDuplicate['newsteamid'][i] = counter

noneDuplicate
```
